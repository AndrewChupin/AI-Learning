```python
import os
import pandas
% matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

HOUSING_PATH = os.path.join("datasets", "housing")


def load_housing_data(housing_path=HOUSING_PATH):
    cvs_path = os.path.join(housing_path, "housing.csv")
    return pandas.read_csv(cvs_path)


frame = load_housing_data()

frame.describe()
```

```python
import numpy as np

def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]


train_set, test_set = split_train_test(frame, 0.2)
print(f"{len(train_set)} train + {len(test_set)} test")
```

```python
from sklearn.model_selection import train_test_split
train_set_std, test_set_std = train_test_split(frame, test_size=0.2, random_state=42)
print(f"{len(train_set_std)} train + {len(test_set_std)} test")
```

```python
frame["income_cat"] = np.ceil(frame["median_income"] / 1.5)
frame["income_cat"].where(frame["income_cat"] < 5, 5.0, inplace=True)
frame.hist(bins=10, figsize=(10, 5), column="income_cat")
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(frame, frame["income_cat"]):
    strat_train_set = frame.loc[train_index]
    strat_test_set = frame.loc[test_index]
frame["income_cat"].value_counts() / len(frame)
```

```python
for set_ in (strat_test_set, strat_train_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

```python
frame = strat_train_set.copy()
frame.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4, s=frame["population"]/100, label="populations",
           figsize=(10, 7), c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True)
plt.legend()
```
